```python
import tkinter as tk
from tkinter import filedialog, messagebox
from stegano import lsb
from PIL import Image, ImageTk

# Function to encode the message in an image
def encode_message():
    file_path = filedialog.askopenfilename(title="Select Image", filetypes=[("Image Files", "*.png")])
    if not file_path:
        return
    
    secret_message = entry_message.get("1.0", tk.END).strip()
    if not secret_message:
        messagebox.showerror("Error", "Message cannot be empty!")
        return

    output_path = filedialog.asksaveasfilename(title="Save Encoded Image", defaultextension=".png",
                                               filetypes=[("Image Files", "*.png")])
    if not output_path:
        return
    
    lsb.hide(file_path, secret_message).save(output_path)
    messagebox.showinfo("Success", "Message encoded successfully!")

# Function to decode the message from an image
def decode_message():
    file_path = filedialog.askopenfilename(title="Select Encoded Image", filetypes=[("Image Files", "*.png")])
    if not file_path:
        return
    
    try:
        hidden_message = lsb.reveal(file_path)
        if hidden_message:
            entry_message.delete("1.0", tk.END)
            entry_message.insert(tk.END, hidden_message)
            messagebox.showinfo("Success", "Message decoded successfully!")
        else:
            messagebox.showerror("Error", "No hidden message found!")
    except Exception as e:
        messagebox.showerror("Error", f"Failed to decode message: {str(e)}")

# Create the main application window
root = tk.Tk()
root.title("Image Steganography")
root.geometry("500x400")

# Widgets
frame = tk.Frame(root, padx=10, pady=10)
frame.pack(pady=20)

label = tk.Label(frame, text="Enter your secret message:")
label.grid(row=0, column=0, sticky="w", pady=5)

entry_message = tk.Text(frame, width=50, height=10)
entry_message.grid(row=1, column=0, pady=5)

btn_encode = tk.Button(root, text="Encode Message", command=encode_message, bg="green", fg="white")
btn_encode.pack(pady=10, ipadx=10)

btn_decode = tk.Button(root, text="Decode Message", command=decode_message, bg="blue", fg="white")
btn_decode.pack(pady=10, ipadx=10)

# Run the application
root.mainloop()
```
